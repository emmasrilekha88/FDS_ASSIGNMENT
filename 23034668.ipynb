```python
#1. Importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#2. Reading the 2020 and 2024 csv file
data_2020 = pd.read_csv(r'C:\Users\dell\Downloads\2020input8.csv', sep='\s+', header=None, names=['Left Interval', 'Right Interval', 'Number of students'])
data_2024 = pd.read_csv(r'C:\Users\dell\Downloads\2024input8.csv', names=['Grade'])
```

```python
#Printing the head of the dataset to check whether the values are assigned with proper column names
print(data_2020.head())
```

```python
print(data_2024.head())
```

```python
#3(a).Calculate mean of the 2020 exam grade file
mean_2020 = np.average(data_2020['Left Interval'], weights=data_2020['Number of students'])

#Calculate of standard deviation of the 2020 exam grade file
std_2020 = np.sqrt(np.average((data_2020['Left Interval'] - mean_2020) ** 2, weights=data_2020['Number of students']))
# rounding upto 2 decimals
mean_2020_rounded = round(mean_2020, 2)
std_2020_rounded = round(std_2020, 2)

print("Mean for 2020 is: ", mean_2020_rounded)
print("Standard deviation for 2020 is: ", std_2020_rounded)

```

```python
#3(b). Calculating the mean of 2024 exam and roundup to 2 decimals.
mean_2024 = np.mean(data2024['Grade'])
mean_2024_rounded = round(mean_2024, 2)
print("mean of 2024 is: ",mean_2024_rounded)

# Calculating the standard deviation for 2024 exam and roundup to 2 decimals.
std_2024 = np.std(data2024['Grade'])
std_2024_rounded = round(std_2024, 2)
print("standard deviation for 2024 is: ", std_2024_rounded)
```

```python
#4. calculating V value
V = round(np.percentile(data2024, 90), 2)
print(V)
```

```python
#5.PLOTTING THE HISTOGRAM
#Plot histogram for 2020_DataSet
plt.figure(figsize=(10, 6))
plt.hist(data_2020['Left Interval'], bins=data_2020['Right Interval'], weights=data_2020['Number of students'], alpha=0.5, label='2020 Exam', color='blue')
plt.text(mean_2020 + 28, 30, f'Mean (2020): {mean_2020:.2f}', color='blue', fontsize=10, ha='center')
plt.text(mean_2020 + 28, 36, f'Standard Deviation (2020): {std_2020:.2f}', color='blue', fontsize=10, ha='center')

# Plot histogram for 2024_DataSet
plt.hist(data_2024['Grade'], bins=range(0, 101, 4), alpha=0.5, label='2024 Exam', color='green')
plt.text(mean_2024 + 29, 28, f'Mean (2024): {mean_2024:.2f}', color='green', fontsize=10, ha='center')
plt.text(mean_2024 + 28, 34, f'Standard Deviation (2024): {std_2024:.2f}', color='green', fontsize=10, ha='center')
plt.text(mean_2024 + 28, 32, f'V: {V:.2f}', color='blue', fontsize=10, ha='center')


#Add registration number as label
plt.text(25, 40, 'Registration Number: 23034668', color='black', fontsize=12, ha='center')

# add the x and y label
plt.xlabel('Exam Grade', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
# add the title and legend
plt.title('Distribution of Exam Grades: 2020 and 2024', fontsize=14)
plt.legend()
# Saving the file
plt.savefig('23034668.png')
plt.show()
```
